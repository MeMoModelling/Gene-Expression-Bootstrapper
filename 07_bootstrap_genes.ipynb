# Bootstrapping Genes with No Expression Values

This tool imputes the expression values for (i) genes used in model but missing expression values, because they could not be mapped to the reference genome and (ii) "unknown genes" created for enzymatic reactions without associated genes.

The output folder from this notebook will serve as the geneExpr folder for FEA.

### Inputs
- **model_pre_filenames**: List of parsed model files from `construct_modularized_community_model` tool
- **mapping_filenames**: List of mapping files to map gene in model to gene in annotation, from `identifiers_mapping` tool
- **combined_geneExpr_filename**: The gene expression values for all species and for all samples

```python
model_pre_filenames = ["data/construct_modularized_community_model/LC_model.xlsx", "data/construct_modularized_community_model/LLm1_model.xlsx", "data/construct_modularized_community_model/LLm2_model.xlsx", "data/construct_modularized_community_model/ST_model.xlsx"] # [1]
mapping_filenames = ["data/identifiers_mapping/LC_mapping.xlsx", "data/identifiers_mapping/LLm1_mapping.xlsx", "data/identifiers_mapping/LLm2_mapping.xlsx", "data/identifiers_mapping/ST_mapping.xlsx"]
combined_geneExpr_filename = "data/bootstrap_genes/combined_geneExpr.csv"
```

### Arguments
- **species_prefixes**: List of species prefixes for the models, in the same order as input files
- **batch_count**: (default: 1000) Number of batch to bootstrap, write each batch to a file

```python
species_prefixes = ["LC", "LLm1", "LLm2", "ST"]
batch_count = 1000
```

### Outputs
- **geneExpr_folder**: Folder containing gene expression files (geneExpr_<1-batch_count>.csv) with bootstrapped values added for unmapped and unknown genes

```python
geneExpr_folder = "data/bootstrap_genes/geneExpr/"
```

### Bootstrapping genes

- For genes with an associated system, a pool of expression values is formed from genes within the same species, same sample, and same system. A value is then randomly selected from this pool.
- For genes without a system, the pool is formed from genes within the same species and same sample (regardless of system), and a value is randomly selected from that broader pool.

```python
from utils.bootstrap_genes import bootstrap_genes

bootstrap_genes(model_pre_filenames, mapping_filenames, species_prefixes, combined_geneExpr_filename, geneExpr_folder, batch_count)
```
